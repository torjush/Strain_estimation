```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.style.use('ggplot')
```

```python
%matplotlib inline
```

```python
results = pd.read_csv('../output/results/double_old_auto/strain_ests.csv')
```

```python
results.columns
```

```python
results['left_diff'] = np.abs(results['left_strain'] - results['left_gt'])
results['right_diff'] = np.abs(results['right_strain'] - results['right_gt'])
```

```python
results.plot(x='left_gt', y='left_strain', kind='scatter')
```

```python
results.plot(x='right_gt', y='right_strain', kind='scatter')
```

```python
lax_res = results[results['view'] == 'lax']
```

```python
four_c_res = results[results['view'] == '4c']
```

```python
two_c_res = results[results['view'] == '2c']
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 7))
left_max = np.nanmax(lax_res[['left_gt', 'left_strain']].values)
left_min = np.nanmin(lax_res[['left_gt', 'left_strain']].values)
x = np.linspace(left_min, left_max)

left_mean = np.nanmean(lax_res['left_diff'])
left_std = np.nanstd(lax_res['left_diff'])
ax[0].plot(x, x, '--', color='grey')
ax[0].scatter(lax_res['left_gt'], lax_res['left_strain'])
ax[0].set_title(f'Anteroseptal, MAD = {left_mean:.2f} (± {left_std:.2f})')

right_max = np.nanmax(lax_res[['right_gt', 'right_strain']].values)
right_min = np.nanmin(lax_res[['right_gt', 'right_strain']].values)
x = np.linspace(right_min, right_max)

right_mean = np.nanmean(lax_res['right_diff'])
right_std = np.nanstd(lax_res['right_diff'])
ax[1].plot(x, x, '--', color='grey')
ax[1].scatter(lax_res['right_gt'], lax_res['right_strain'])
ax[1].set_title(f'Posterior, MAD = {right_mean:.2f} (± {right_std:.2f})')
```
